```python
# Generate small subset to test on to convert the MADW to utf-8...lets do 1GB
import codecs
BLOCKSIZE = 1048576 # or some other, desired size in bytes
sourceFileName = r'/Users/mh302/Desktop/DoD_REF24.csv'
targetFileName = r'/Users/mh302/Desktop/DoD_REF24_utf8_subset.csv'
with codecs.open(sourceFileName, "r", "latin1") as sourceFile:
    with codecs.open(targetFileName, "w", "utf-8") as targetFile:
        n = 0
        while n < 1000:
            contents = sourceFile.read(BLOCKSIZE)
            if not contents:
                break
            targetFile.write(contents)
            n+=1
```

```python
# this converted the MADW file from latin1 to utf-8, since DASK only works on utf-8 (can't pass encoding arg)
import codecs
BLOCKSIZE = 1048576 # or some other, desired size in bytes
sourceFileName = r'/Users/mh302/Desktop/DoD_REF24.csv'
targetFileName = r'/Users/mh302/Desktop/DoD_REF24_full_utf8.csv'
with codecs.open(sourceFileName, "r", "latin1") as sourceFile:
    with codecs.open(targetFileName, "w", "utf-8") as targetFile:
        while True:
            contents = sourceFile.read(BLOCKSIZE)
            if not contents:
                break
            targetFile.write(contents)
```
